```python
def parse_chunk(chunk):
    chunk = chunk.strip()

    values = chunk.split(",")

    name = values[0]
    followers = int(values[1])
    following = int(values[2])
    post_no = int(values[3])
    bio = values[4]

    return {
        "name": name,
        "followers": followers,
        "following": following,
        "post_no": post_no,
        "bio": bio
    }

file = open("openai_instagram_followers_10000_synthetic.csv","r")

data = file.readlines()[1:]   # skip header

for line in data[:5]:   # first 5 rows
    dicto = parse_chunk(line)
    print(dicto)

file.close()
```
